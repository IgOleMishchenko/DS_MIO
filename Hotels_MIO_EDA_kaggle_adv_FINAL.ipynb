```python
import pandas as pd
```

```python
hotels = pd.read_csv('data/hotels.csv')
hotels.head()

tags = ()

def split(row):
    array = row.replace('[]\'', '').split(',')
    for value in array:
        tags.add(value)
```

```python
hotels.info()
```

```python
#Удаляем колонки, которые не могут оказать влияние на предсказание рейтинга отеля 
hotels_cleaned = hotels.drop(columns=['hotel_address', 'hotel_name', 'negative_review', 'positive_review', 'tags', 'lat', 'lng'], axis=1)
```

```python
hotels_cleaned.head()
```

```python
#Создаем новый признак. Создаем функцию, которая преобразовывает колонку "review_date" из  даты в сезон
def getSeasonFromDate(date):
    month = int(date.split('/')[0])
    if (month >= 3 and month <= 5):
        return 1
    elif (month >= 6 and month <= 8):
        return 2
    elif (month >= 9 and month <= 11):
        return 3
    else:
        return 4
    
hotels_cleaned['season'] = hotels_cleaned['review_date'].apply(getSeasonFromDate)
hotels_cleaned.head()
```

```python
#Удаляем столбец "review_date", так как мы создали новый признак - "Season" (Сезон - в какое время года сделан отзыв)
hotels_cleaned.drop('review_date', axis=1)
```

```python
hotels_cleaned['days_from_review'] = hotels_cleaned['days_since_review'].apply(lambda x : int(x.split(' ')[0]))
hotels_cleaned = hotels_cleaned.drop(columns=['days_since_review', 'review_date'], axis=1)
hotels_cleaned.head()
```

```python
hotels_cleaned_f = hotels_cleaned.drop(columns=['days_from_review', 'reviewer_nationality'], axis=1)
hotels_cleaned_f.head()
```

```python
hotels_cleaned_f.info()
```

```python
# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели  
# Х — данные с информацией об отелях, у — целевая переменная (рейтинги отелей)  
X = hotels_cleaned_f.drop(['reviewer_score'], axis = 1)  
y = hotels_cleaned_f['reviewer_score']  
      
# Загружаем специальный инструмент для разбивки:  
from sklearn.model_selection import train_test_split  
      
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" — для тестирования.  
# Для тестирования мы будем использовать 25 % от исходного датасета.  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Импортируем необходимые библиотеки:  
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели  
from sklearn import metrics # инструменты для оценки точности модели  
      
# Создаём модель  
regr = RandomForestRegressor(n_estimators=100)  
      
# Обучаем модель на тестовом наборе данных  
regr.fit(X_train, y_train)  
      
# Используем обученную модель для предсказания рейтинга отелей в тестовой выборке.  
# Предсказанные значения записываем в переменную y_pred  
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test) и смотрим, насколько они отличаются  
# Метрика называется Mean Absolute Percentage Error (MAPE) и показывает среднюю абсолютную процентную ошибку предсказанных значений в отношении фактических.  
print('MAPE:', metrics.mean_absolute_percentage_error(y_test, y_pred))
```
